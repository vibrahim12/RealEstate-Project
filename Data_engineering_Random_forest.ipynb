```python
# # Starting dataFrame. df_combined has all the ETL we have done in the past
data_type_mapping = {"SUBMARKET_CODE": "object", "PROPERTY_ZIPCODE": "object", "PROPERTY_PHONE": "object", "census_tract": "object", "county_code": "object",
                     "state_code": "object"}
df_combined_old = pd.read_csv("df_combined.csv", dtype=data_type_mapping)
df_combined_old = df_combined_old[["PROPERTY_ID", "census_tract"]]
df_combined = pd.read_csv("df_combined_target.csv", dtype=data_type_mapping)
df_combined = df_combined.merge(df_combined_old, how="left", on="PROPERTY_ID")
df_combined["census_tract"] = df_combined["census_tract_y"]
df_combined = df_combined.drop(["census_tract_x", "census_tract_y"], axis=1)

# # df_future_supply has all the property that were to be built in the future. 
df_future_supply = pd.read_csv("df_future_supply.csv", dtype=data_type_mapping)

# # Drop unnecessary columns
df_combined = df_combined.drop(["Unnamed: 0", "Unnamed: 0.1"], axis=1)
df_future_supply = df_future_supply.drop(["Unnamed: 0"], axis=1)


def update_census_tract(row):
  census_tract = row["census_tract"]
  pattern = re.compile("^\d{4}\.\d{2}$")
  ct_list = census_tract.split(".")
  if pattern.match(census_tract):
    return ct_list[0] + ct_list[-1]
  
  if len(ct_list[0]) < 4:
    prefix = "0" * (4 - len(ct_list[0])) + ct_list[0]
  else:
    prefix = ct_list[0]
  
  if len(ct_list[-1]) < 2:
    suffix = ct_list[-1] + "0" * (2 - len(ct_list[-1]))
  else:
    suffix = ct_list[-1]
  return prefix + suffix

def update_county_code(row):
  county_code = row["county_code"]
  pattern = re.compile("^\d{3}$")
  if pattern.match(county_code):
    return county_code
  if len(county_code) < 3:
    county_code = "0" * (3 - len(county_code)) + county_code
    return county_code

def update_state_code(row):
  state_code = row["state_code"]
  pattern = re.compile("^\d{2}$")
  if pattern.match(state_code):
    return state_code
  if len(state_code) < 2:
    state_code = "0" * (2 - len(state_code)) + state_code
    return state_code

def update_age_of_the_property(row):
    age_of_the_property = row["age_of_the_property"]
    return age_of_the_property.split(" ")[0]

df_combined["census_tract"] = df_combined.apply(lambda row: update_census_tract(row), axis=1)
df_combined["county_code"] = df_combined.apply(lambda row: update_county_code(row), axis=1)
df_combined["state_code"] = df_combined.apply(lambda row: update_state_code(row), axis=1)
# df_combined["age_of_the_property"] = df_combined.apply(lambda row: update_age_of_the_property(row), axis=1)

df_future_supply["census_tract"] = df_future_supply.apply(lambda row: update_census_tract(row), axis=1)
df_future_supply["county_code"] = df_future_supply.apply(lambda row: update_county_code(row), axis=1)
df_future_supply["state_code"] = df_future_supply.apply(lambda row: update_state_code(row), axis=1)

# # Sanity check, no rows should show. 
print(df_combined[~df_combined.census_tract.str.match("^\d{6}$")])
print(df_combined[~df_combined.county_code.str.match("^\d{3}$")])
print(df_combined[~df_combined.state_code.str.match("^\d{2}$")])
```

```python
import numpy as np

df_combined_grouped = df_combined_ungrouped_final
target= np.array(df_combined_grouped['rent_per_sqft'])
features = df_combined_grouped.drop('rent_per_sqft', axis = 1)
feature_list = list(features.columns)
features = np.array(features)
X_train1, X_test1, y_train1, y_test1 = \
    train_test_split(features, target,
                     test_size=0.5,
                     random_state=2)

## RANDOM FOREST - KFOLD AND MODEL 


from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train1, y_train1)
```
